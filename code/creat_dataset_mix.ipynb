```python
import pandas as pd
```

```python
happy_df = pd.read_csv("Happy_song_playlist.csv")
```

```python
happy_df['label'] = 1
happy_df.tail()
print(happy_df.shape)
```

```python
sad_df = pd.read_csv("Sad_song_playlist.csv")
```

```python
sad_df.head()
sad_df['label']=0
sad_df.tail()
print(sad_df.shape)
```

#### Mixing and shuffling happy and sad

```python
mix_df = pd.concat([sad_df, happy_df])
print(mix_df.shape)
mix_df.head()
```

```python
mix_df = mix_df.sample(frac=1)
mix_df.reset_index(drop=True)
```

```python
mix_df.head()
```

```python
mix_df.to_csv("Mix_playlist.csv")
```
